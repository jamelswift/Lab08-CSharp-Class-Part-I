## คำถาม

1. Field ในภาษา C# คืออะไร


*คำตอบ*Field คือ ตัวแปรของคลาส ที่ใช้เก็บข้อมูลของอ็อบเจกต์

2. Field มีหน้าที่อะไรใน class 


*คำตอบ*
หน้าที่หลักของ Field
1.เก็บสถานะของอ็อบเจกต์ (Store Object State)
Field ใช้สำหรับเก็บค่าของ แต่ละอ็อบเจกต์ ที่สร้างจากคลาสนั้นๆ
เช่น คลาส Person อาจมี Field name และ age เพื่อเก็บข้อมูลของแต่ละบุคคล

2.ใช้เป็นค่าพื้นฐานสำหรับการคำนวณหรือการทำงานในคลาส
Field อาจถูกใช้เพื่อเก็บค่าที่นำไปคำนวณใน Method อื่นๆ
เช่น Field salary ใช้ในการคำนวณโบนัสหรือภาษีของพนักงาน

3.ใช้เป็นตัวนับหรือค่าร่วมระหว่างอ็อบเจกต์ (Shared Data - Static Field)
Field สามารถเป็น static เพื่อให้ ใช้ร่วมกันทุกอ็อบเจกต์ ของคลาส
เช่น คลาส Employee อาจมี Static Field employeeCount เพื่อเก็บจำนวนพนักงานทั้งหมดที่ถูกสร้างขึ้น

4.ช่วยให้คลาสมีข้อมูลคงที่หรือค่าเริ่มต้น (Default or Constant Data)
ใช้ readonly หรือ const สำหรับค่าที่ไม่ควรเปลี่ยนแปลงหลังจากกำหนด
เช่น const double PI = 3.14159; ใช้เป็นค่าคงที่ของคลาส

3. การประกาศ Field ทำได้อย่างไร

*คำตอบ*
การประกาศ Field ใน C# คือการสร้าง ตัวแปรของคลาส (class) หรือโครงสร้าง (struct) เพื่อใช้เก็บค่าหรือสถานะของอ็อบเจกต์ (Object State)
Field สามารถมีตัวกำหนดการเข้าถึง (Access Modifiers) เช่น public, private, protected, และสามารถกำหนดให้เป็น static, readonly, หรือ const ได้

4. Implicit fields initialization  คืออะไร

*คำตอบ*
Implicit Fields Initialization คือ ค่าตั้งต้น (default value) ที่ C# กำหนดให้ Field ของคลาสหรือโครงสร้าง (class หรือ struct) โดยอัตโนมัติ หากไม่ได้กำหนดค่าเริ่มต้นให้เอง

5. Explicit field initialization คืออะไร



*คำตอบ*
Explicit Field Initialization คือการ กำหนดค่าเริ่มต้นให้ Field โดยตรง ขณะประกาศ Field ในคลาส (class) หรือโครงสร้าง (struct) แทนที่จะปล่อยให้ C# กำหนดค่าให้โดยอัตโนมัติ (Implicit Initialization)

6. Implicit fields initialization  และ Explicit field initialization แตกต่างกันอย่างไร



*คำตอบ*
Implicit Field Initialization: C# กำหนดค่าเริ่มต้นให้กับ Field โดยอัตโนมัติ ตามประเภทของข้อมูล

Explicit Field Initialization: นักพัฒนากำหนดค่าเริ่มต้นให้ Field ด้วยตนเอง เพื่อใช้ค่าเริ่มต้นที่ต้องการ
